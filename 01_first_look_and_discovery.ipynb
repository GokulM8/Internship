## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
# Your discovery here
import pandas as pd
import matplotlib.pyplot as plt

#loading the dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# keeping only pages with enough impressions for reliable CTR
visible = df[df["impressions_90d"] >= 100]

#Aveerage CTR by ranking tier and content type
ctr_analysis = (
    visible.groupby(["position_tier", "content_type"])["ctr"]
    .agg(["mean", "count"])
    .reset_index()
)

#remaming the columns
ctr_analysis.columns = ["position_tier", "content_type", "avg_ctr", "pages"]

#sorting within each position tier
ctr_analysis = ctr_analysis.sort_values(by=["position_tier", "avg_ctr"], ascending=[True, False])

#displaying the results
print(ctr_analysis)

#Show the lowest CTR content type in eaxh position tier
worst_ctr = ctr_analysis.loc[ctr_analysis.groupby("position_tier")["avg_ctr"].idxmin()]
print(worst_ctr)

ctr_analysis_filtered = ctr_analysis[ctr_analysis["pages"] >= 30]

print(ctr_analysis_filtered)

pivot = ctr_analysis.pivot(
    index="position_tier",
    columns="content_type",
    values="avg_ctr"
)

pivot.plot(kind="bar", figsize=(10,5))

plt.title("Average CTR by Position Tier and Content Type")
plt.ylabel("Average CTR")
plt.xlabel("Position Tier")
plt.xticks(rotation=0)
plt.legend(title="Content Type")
plt.show()
```

## Observation

I analyzed the average Click-Through Rate (CTR) for different content types within the same search position tiers, considering only groups with a sufficient number of pages. I observed that comparison articles consistently had lower average CTR than keyword articles across the Page 1, Page 3–5, and Striking position tiers. While feedly articles showed higher CTR in some tiers, they were represented by a much smaller number of pages, so those results should be interpreted cautiously. Overall, this suggests that content type may influence user engagement even when pages have similar search rankings, although this is an observation from the anonymized dataset rather than a general rule.
